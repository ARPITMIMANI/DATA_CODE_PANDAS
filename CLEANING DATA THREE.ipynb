```python
# SET "Duration" = 45 IN ROW 7 .

import pandas as pd
b = pd.read_excel("DIRTY_DATA.xlsx")
b.loc [ 7 , "Duration" ] = 45
print ( b.to_string() )
```

```python
'''
FOR LARGER DATA SET LOOP THROUGH ALL THE VALUES IN THE "Duration" COLUMN ,
IF THE VALUE IS HIGHER THAN 120 THEN SET IT TO 120 .
''' 

import pandas as pd
c = pd.read_excel("DIRTY_DATA.xlsx")
for x in c.index :
    if c.loc [ x , "Duration" ] > 120 :
       c.loc [ x , "Duration" ] = 120 
print ( c.to_string() )    
```

```python
# FOR A LARGER DATA SET REMOVE THE ROWS FOR WRONG DATA .

import pandas as pd
d = pd.read_excel("DIRTY_DATA.xlsx")
for y in d.index :
    if d.loc [ y , "Duration" ] > 120 :
        d.drop ( y , inplace=True )
print ( d.to_string() )        
```
