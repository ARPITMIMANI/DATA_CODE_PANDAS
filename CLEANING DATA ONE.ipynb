```python
# CLEANING DATA : IT MEANS FIXING THE BAD DATA IN THE DATA SET .
# BAD DATA COULD BE EMPTY CELL , DUPLICATE DATA , WRONG DATA AND DATA IN A WRONG FORMAT .

# EMPTY CELL : IT WILL GIVE WRONG RESULTS ALWAYS . WE HAVE TO REMOVE THE ROWS THAT CONTAINS BAD DATA .

# LOAD THE DATA FROM THE EXCEL FILE "DIRTY_DATA.xlsx" INTO DATAFRAME .

import pandas as pd
df = pd.read_excel("DIRTY_DATA.xlsx")
print ( df.to_string() )
```

```python
'''
LOAD THE DATA FROM THE EXCEL FILE "DIRTY_DATA.xlsx" INTO DATAFRAME WITH NO EMPTY CELLS 
WITHOUT CHANGING ORIGINAL DATAFRAME .
''' 

# dropna() IS USED TO REMOVE THE ROWS THAT CONTAINS NULL (NaN) VALUES .

import pandas as pd
df = pd.read_excel("DIRTY_DATA.xlsx")
aa = df.dropna()
print ( aa.to_string() )
```

```python
'''
LOAD THE DATA FROM THE EXCEL FILE "DIRTY_DATA.xlsx" INTO DATAFRAME WITH NO EMPTY CELLS 
WITH CHANGING ORIGINAL DATAFRAME .
''' 

import pandas as pd
bb = pd.read_excel("DIRTY_DATA.xlsx")
bb.dropna ( inplace=True )
print ( bb.to_string() )
```

```python
'''
LOAD THE DATA FROM THE EXCEL FILE "DIRTY_DATA.xlsx" INTO DATAFRAME WITH NO EMPTY CELLS 
WITHOUT CHANGING ORIGINAL DATAFRAME .
''' 

# fillna() IS USED REPLACE THE EMPTY CELL OR NULL VALUES WITH A VALUE

import pandas as pd
df = pd.read_excel("DIRTY_DATA.xlsx")
cc = df.fillna(130)
print ( cc.to_string() )
```

```python
'''
LOAD THE DATA FROM THE EXCEL FILE "DIRTY_DATA.xlsx" INTO DATAFRAME WITH NO EMPTY CELLS 
WITH CHANGING ORIGINAL DATAFRAME .
''' 

import pandas as pd
dd = pd.read_excel("DIRTY_DATA.xlsx")
dd.fillna ( 130 , inplace=True )
print ( dd.to_string() )
```

```python
# TO REPLACE ONLY THE EMPTY VALUE FOR PARTICULAR COLUMN WE NEED TO SPECIFIY THE COLUMN NAME .

import pandas as pd
ee = pd.read_excel("DIRTY_DATA.xlsx")
ee["Calories"].fillna ( 130 , inplace=True )
print ( ee.to_string() )
```

```python
import pandas as pd
ee = pd.read_excel("DIRTY_DATA.xlsx")
ee.Calories.fillna ( 130 , inplace=True )
print ( ee.to_string() )
```

```python
# CALCULATE THE mean() AND REPLACE THE EMPTY CELLS WITH IT .

import pandas as pd
ff = pd.read_excel("DIRTY_DATA.xlsx")
xx = ff["Calories"].mean()
ff.fillna ( xx , inplace=True )
print ( ff.to_string() )
```

```python
# CALCULATE THE median() AND REPLACE THE EMPTY CELLS WITH IT .

import pandas as pd
ff = pd.read_excel("DIRTY_DATA.xlsx")
yy = ff["Calories"].median()
ff.fillna ( yy , inplace=True )
print ( ff.to_string() )
```

```python
# CALCULATE THE mode() AND REPLACE THE EMPTY CELLS WITH IT .

import pandas as pd
ff = pd.read_excel("DIRTY_DATA.xlsx")
zz = ff["Calories"].mode()[0]
ff.fillna ( zz , inplace=True )
print ( ff.to_string() )
```
